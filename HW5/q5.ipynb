```python
import pandas as pd 
```

```python
df = pd.read_csv("iris_dataset.csv")
df
```

```python
sepal = df[["sepal length (cm)" , "sepal width (cm)"]]
target = df[["target"]]

ptest = sepal.iloc[149]
ptest_target = target.iloc[149]

sepal_train = sepal.iloc [ : 149 ] 
target_train = target.iloc [ : 149 ]

print("Test sample: \n", ptest)
print("Real class: \n", ptest_target)
```

```python
import numpy as np

distances = []

for i in range(len(sepal_train)):
    point = sepal_train.iloc[i]
    dist = np.sqrt((point[0] - ptest[0])**2 + (point[1] - ptest[1])**2)
    distances.append(dist)

min_index = np.argmin(distances) 
predicted_class = target_train.iloc[min_index][0]

print(" Ptest class:", predicted_class)
```

```python
k = 3
nearest_indices = np.argsort(distances)[:k] 

nearest_classes = target_train.iloc[nearest_indices].values.flatten()

values, counts = np.unique(nearest_classes, return_counts=True)
predicted_class_k3 = values[np.argmax(counts)]

print("Classes 3 nearest neighbors:", nearest_classes)
print("Predicted class k=3:", predicted_class_k3)
```
